# 2 Different Simulations Below:
- These two simulations is a particle simulation and dice simulation

```python
# Set the initial conditions
time = 0
position = 0
velocity = 10

# Set the time step (in seconds)
dt = 0.1

# Set the simulation duration (in seconds)
duration = 10

while time <= duration:
    # Update the position and time based on the current velocity
    position += velocity * dt
    time += dt
    
    # Print the current state
    print("Time:", time)
    print("Position:", position)
    print("Velocity:", velocity)
```

```python
import random

# Roll a six-sided dice and return the result
def roll_dice():
  return random.randint(1, 6)

# Prompt the user for the number of dice rolls
num_rolls = int(input("Enter the number of times you want to roll the dice: "))

# Simulate rolling the dice the specified number of times and print the results
for i in range(num_rolls):
  result = roll_dice()
  print(f"Roll {i + 1}: {result}")
```
